### Load packages

```python
import librosa
import librosa.display as display
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import HTML

%matplotlib inline
```

### Declare constants

```python
RATE = 16000
N_MFCC = 13
COL_SIZE = 30
```

### Load wav file & downsample

```python
wav, original_rate = librosa.load('audio/mandarin40.wav')
wav = librosa.core.resample(y=wav,orig_sr=original_rate,target_sr=RATE, scale=True)
```

### Plot wave form

```python
plt.figure(figsize=(11.5, 8))
plt.title('Wave form')
plt.ylabel('Amplitude')
plt.xlabel('Time')
plt.plot(wav)
```

### Convert wav to Mel Spectrogram
Only used as an example

```python
melspec = librosa.feature.melspectrogram(y=wav, sr=RATE, n_fft=2048, hop_length=512, power=2.0)
```

```python
plt.figure(figsize=(15, 8))
librosa.display.specshow(librosa.power_to_db(melspec,ref=np.max),y_axis='mel', fmax=12000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Smfccs = librosa.feature.mfcc(y=wav, sr=RATE, n_mfcc=N_MFCC)pectrogram')
plt.tight_layout()
```

### Convert wav to Mel Frequencies Cepstral Coefficients
Each column is around 20ms

```python
mfccs = librosa.feature.mfcc(y=wav, sr=RATE, n_mfcc=N_MFCC)
```

```python
plt.figure(figsize=(15, 8))
display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```

```python
plt.figure(figsize=(15, 8))
display.specshow(mfccs[:,500:500+COL_SIZE], x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
```
